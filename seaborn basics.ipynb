# Seaborn plotting

## Univariate

### 1) Histogram

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=np.random.randn(1000)
```

```python
sns.histplot(data,bins=30,color="blue")
plt.title("Histogram")
plt.xlabel("value")
plt.ylabel("frequency")
plt.show()
```

### 2)KDE plot (kernel density estimate)

```python
sns.kdeplot(data,shade=True,color="purple")
plt.title("KDE")
plt.xlabel("value")
plt.ylabel("frequency")
plt.show()
```

### 3) Box plot

```python
sns.boxplot(x=data,color="cyan")
plt.title("Boxplot")
plt.xlabel("value")
plt.show()
```

### 4) violin plot

```python
sns.violinplot(x=data,color="magenta")
plt.title("Violin plot")
plt.xlabel("value")
plt.show()
```

## Bivariate plots

### 1)scatterplot

```python
x=np.random.rand(100)
y=np.random.rand(100)
sns.scatterplot(x=x,y=y,color='green')
plt.title("scatter plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

### 2) lineplot

```python
x=np.linspace(0,10,100)
y=np.sin(x)
sns.lineplot(x=x,y=y,color="black")
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y=axis")
plt.show()
```

### 3) Heat map

```python
data=np.random.rand(10,10)
sns.heatmap(data,annot=True,cmap="coolwarm")
plt.title("heatmap")
plt.show()
```

### 4) Barplot

```python
categories=['a','b','c']
values=[10,20,15]
sns.barplot(x=categories,y=values,palette="muted")
plt.title("bar plot")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

### 5) pair plot

```python
import pandas as pd
data=pd.DataFrame({'x':np.random.rand(50),"y":np.random.rand(50),"z":np.random.rand(50)})
sns.pairplot(data)
plt.suptitle("pair plot",y=1.02)
plt.show()
```
